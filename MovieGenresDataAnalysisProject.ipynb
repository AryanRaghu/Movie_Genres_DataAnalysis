```python
import pandas as pd
movies=pd.read_csv('/content/imdb_movies.csv')
movies
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option('display.max.columns',22)
```

```python
movies.head()
```

```python
movies[movies.duplicated()]
```

```python
 movies.drop_duplicates(inplace=True)  #removing duplicated values
movies[movies.duplicated()]
```

```python
movies.dropna(subset=['genres'],inplace=True)
```

```python
movies.info()
```

```python
movies['profit']=movies['revenue']-movies['budget']
```

```python
movies_genre=movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series,DataFrame
```

```python
split=movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name='genres_split'
del movies_genre['genres']
movies_genre=movies_genre.join(split)
```

```python
movies_genre
```

# **Research Questions (Q)**

#Q1) Which genres are the most common (number of movies made)?

```python
genres_count=pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending=False)
genres_count
```

```python
genres_count['original_title'].plot.pie(title='Movies Per Genre in %',autopct='%1.1f%%',figsize=(10,10))
```

```python
genres_count['original_title'].plot.barh(title='Movies Per Genre',color='DarkBlue',figsize=(10,9))
```

# Q2) Which genres have high avg budget and revenue

```python
numeric_columns = movies_genre.select_dtypes(include=['number']).columns
genres_avg = movies_genre.groupby('genres_split')[numeric_columns].mean()
pd.options.display.float_format= '{:2f}'.format
print(genres_avg)
```

```python
genres_avg.sort_values('budget',ascending=False,inplace=True)
```

```python
genres_avg[['budget','revenue']].plot.barh(title='Budget and Revenue by Genre',color=('DarkBlue','c'),figsize=(10,9))
```

# Q3) Which genres have high avg. Profit?

```python
genres_avg.sort_values('profit',ascending=True,inplace=True)
```

```python
genres_avg['profit'].plot.barh(title='Profit by Genre',color='DarkBlue',figsize=(10,9))
```

# Q4) Which genres have high avg. Popularity?

```python
genres_avg.sort_values('popularity',ascending=True,inplace=True)
```

```python
genres_avg['popularity'].plot.barh(title='Popularity of the Genre',color='DarkBlue',figsize=(10,9))
```

# Research Hypothesis (H)

# Q1) The best movies according to vote avg return high profit and revenue

```python
movies.drop_duplicates(inplace=True)
movies['profit']=movies['revenue']-movies['budget']
movies_genre=movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
movies_counted=movies_genre[movies_genre['vote_count']>=50]
# Select only numeric columns from the DataFrame
numeric_columns = movies_counted.select_dtypes(include=['number'])

# Calculate the Spearman correlation matrix
correlation_matrix = numeric_columns.corr(method='spearman')

# Print the correlation matrix
print(correlation_matrix)
```

```python
movies_counted.plot.scatter(x='vote_average',y='profit')
```

```python
import seaborn as sns
sns.regplot(x='vote_average',y='profit',data=movies_counted,line_kws={'color':'red'})
```

# Q2) The best movies according to popularity return high profit and revenue

```python
numeric_columns = movies_counted.select_dtypes(include=['number'])

# Calculate the Spearman correlation matrix
correlation_matrix = numeric_columns.corr(method='spearman')

# Print the correlation matrix
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
sns.regplot(x='popularity',y='revenue',data=movies_counted,line_kws={'color':'red'})
plt.figure(figsize=(20,5))
plt.show()
```

# Q3) Highly budgeted movies return high profit

```python
sns.regplot(x='budget',y='profit',data=movies_counted,line_kws={'color':'red'})
plt.figure(figsize=(10,5))
plt.show()
```

```python
sns.regplot(x='budget',y='popularity',data=movies_counted,line_kws={'color':'red'})
plt.figure(figsize=(10,5))
plt.show()
```
